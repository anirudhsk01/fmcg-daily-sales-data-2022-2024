### Key Features

1. **Date**  
   - **Type:** Date (object → convertible to datetime)  
   - **Description:** The transaction date of the sales record.  
   - **Relevance:** Enables time-series analysis such as trends, seasonality, and forecasting.

2. **SKU (Stock Keeping Unit)**  
   - **Type:** Categorical (string)  
   - **Description:** Unique identifier for each product item.  
   - **Relevance:** Allows tracking of product-level performance. 

3. **Brand**  
   - **Type:** Categorical (string)  
   - **Description:** Name of the product’s brand.  
   - **Relevance:** Helps in brand-level sales comparison and loyalty analysis. 

4. **Segment**  
   - **Type:** Categorical (string)  
   - **Description:** Market segment or target group the product belongs to.  
   - **Relevance:** Facilitates customer segmentation and targeted strategies. 

5. **Category**  
   - **Type:** Categorical (string)  
   - **Description:** Higher-level product category (e.g., beverages, snacks).  
   - **Relevance:** Useful for category-level performance monitoring.

6. **Channel**  
   - **Type:** Categorical (string)  
   - **Description:** Sales channel (e.g., retail, wholesale, online).  
   - **Relevance:** Assists in analyzing channel effectiveness. 

7. **Region**  
   - **Type:** Categorical (string)  
   - **Description:** Geographical region of sales.  
   - **Relevance:** Supports regional demand and distribution analysis.

8. **Pack_Type**  
   - **Type:** Categorical (string)  
   - **Description:** Type of product packaging (e.g., bottle, box, pouch).  
   - **Relevance:** Helps understand consumer preference for packaging formats.

9. **Price_Unit**  
   - **Type:** Numerical (float)

10. **Promotion_Flag**  
   - **Type:** Numerical (integer: 0/1)

11. **Delivery_Days**

12. **Stock_Available**

   - **Description:** Quantity of stock available at the time of sale.

13. **Delivered_Qty**  
   - **Type:** Numerical (integer)

   - **Relevance:** Important for tracking supply fulfillment and logistics.

14. **Units_Sold**

## 9. DATA PRE-PROCESSING

To prepare the dataset for analysis, the following steps were performed:  

- **Handling Missing Values:** Checked missing percentages, filled categorical values with mode, numerical with median, and dropped rows with missing dates.  
- **Removing Duplicates:** Identified and removed duplicate rows; final dataset shape confirmed.  
- **Correcting Data Types:** Converted columns to proper formats (e.g., `date` → datetime, numeric fields → numeric).  
- **Creating Derived Columns:** Added new features like `Revenue (Quantity × Unit Price)` and extracted `Year`, `Month`, and `Day` from date.  
- **Filtering & Aggregating Data:** Filtered invalid/zero sales and summarized data by month and category.  

These steps ensured data quality, consistency, and readiness for further analysis

```python
# 1. Handling missing values

print("Missing values before handling:\n")
print(df.isnull().sum())
```

```python
# 2. Percentage of missing values per column

print("\n Percentage of missing values per column:\n")
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Separate categorical and numerical
# 3. Select categorical columns (object + category)

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:\n", categorical_cols)
```

```python
# 4. Select numerical columns (int + float)

numerical_cols = df.select_dtypes(include=['int64','float64']).columns
print("Numerical Columns:\n", numerical_cols)
```

```python
# Fill missing values
# 5. Handle categorical columns

for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        print(f" Filling missing values in '{col}' with mode: {df[col].mode()[0]}")
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        print(f" No missing values found in '{col}'")
```

```python
# 6. Handle numerical columns

for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        print(f" Filling missing values in '{col}' with median: {df[col].median()}")
        df[col].fillna(df[col].median(), inplace=True)
    else:
        print(f" No missing values found in '{col}'")
```

```python
# 7. Drop rows if 'date' is missing

if df['date'].isnull().sum() > 0:
    print(f" Dropping {df['date'].isnull().sum()} rows with missing 'date'")
    df.dropna(subset=['date'], inplace=True)
    
print("\n Missing values after handling:\n")
print(df.isnull().sum())
```

```python
# Removing Duplicates
# 8. Check number of duplicate rows

duplicates = df.duplicated().sum()
print(f" Number of duplicate rows before removal: {duplicates}")
```

```python
# 9. Remove duplicates

df.drop_duplicates(inplace=True)
print(" Rows after removing duplicates:", df.shape[0])
print(" Duplicates removed:", df.shape[0] - df.drop_duplicates().shape[0])
```

```python
# 10. Final dataset shape after removing duplicates

print(f" Dataset shape after duplicate removal: {df.shape}")
```

```python
# Correcting Data Types
# 11. Convert 'date' column to datetime

if df['date'].dtype != 'datetime64[ns]':
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    print(" 'date' column converted to datetime")
else:
    print(" 'date' column is already datetime:", df['date'].dtype)
```

```python
# 12. Convert 'promotion_flag' to categorical (0/1)

if df['promotion_flag'].dtype != 'category':
    df['promotion_flag'] = df['promotion_flag'].astype('category')
    print(" Converted 'promotion_flag' column to categorical")
else:
    print(" 'promotion_flag' is already categorical")
```

```python
# 13. Convert product/region related columns to categorical

categorical_columns = ['sku', 'brand', 'segment', 'category', 'channel', 'region', 'pack_type']

for col in categorical_columns:
    if df[col].dtype != 'category':
        df[col] = df[col].astype('category')
        print(f" Converted '{col}' column to categorical")
    else:
        print(f" '{col}' column is already categorical")
```

```python
# 14. Final Check

print("\n Final Data Types After Conversion:\n")
print(df.dtypes)
```

```python
# Creating Derived Columns
# 15. Total Revenue = units_sold × price per unit

df['revenue'] = df['units_sold'] * df['price_unit']
print("Created 'revenue' column (units_sold × price_unit)")
```

```python
# 16. Delivery Efficiency = delivered_qty / (units_sold + 1)

df['delivery_efficiency'] = df['delivered_qty'] / (df['units_sold'] + 1)
print(" Created 'delivery_efficiency' column (delivered_qty ÷ units_sold)")
```

```python
# 17. Stock Utilization = units_sold ÷ (stock_available + 1)

df['stock_utilization'] = df['units_sold'] / (df['stock_available'] + 1)
print(" Created 'stock_utilization' column (units_sold ÷ stock_available)")
```

```python
# 18. Month and Year from Date (for time-based analysis)

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
print(" Extracted 'year' and 'month' from 'date' column")
```

```python
# 19. Promotion Label (Yes/No) for better readability

df['promotion_label'] = df['promotion_flag'].map({1: "Yes", 0: "No"})
print(" Created 'promotion_label' column (Yes/No)")
```

```python
# 20. Show a preview of new columns

print("\n Preview of Derived Columns:\n")
print(df[['date','units_sold','price_unit','revenue','delivery_efficiency','stock_utilization','year','month','promotion_label']].head())
```

```python
# Filtering and Aggregating Data
# --- Filtering ---
# 21. Filter only rows where promotion was active

promo_sales = df[df['promotion_flag'] == 1]
print(" Filtered rows where promotion was active")
print(promo_sales.head(), "\n")
```

```python
# 22. Filter data for year 2024 only

sales_2024 = df[df['year'] == 2024]
print(" Filtered rows for year 2024")
print(sales_2024.head(), "\n")
```

```python
# --- Aggregation ---
# 23. Total revenue by product

print(df.columns)
revenue_by_sku = df.groupby('sku')['revenue'].sum().reset_index()
print(revenue_by_sku.head())
```

```python
# 24. Average delivery efficiency by channel

avg_efficiency_by_channel = df.groupby('channel')['delivery_efficiency'].mean().reset_index()
print(" Aggregated average delivery efficiency by channel")
print(avg_efficiency_by_channel.head(), "\n")
```

```python
# 25. Monthly revenue trend

monthly_revenue = df.groupby(['year','month'])['revenue'].sum().reset_index()
print("Aggregated monthly revenue trend")
print(monthly_revenue.head(), "\n")
```

## 10. EXPLORATORY DATA ANALYSIS (EDA)

In this step, we aim to explore the dataset to uncover insights, patterns, and relationships.  

- **Dataset Overview:** Checked the shape, column names, and data types to understand the structure.  
- **Univariate Analysis:** Explored single features using frequency counts, distributions, and summary statistics (mean, median, std).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 26. Univariate Analysis
# --- Numerical features ---
print("\n--- Statistical Summary of Numerical Columns ---")
print(df.describe())

# Distribution of revenue
plt.figure(figsize=(8,5))
sns.histplot(df['revenue'], bins=30, kde=True)
plt.title("Revenue Distribution")
plt.show()

# Units sold
plt.figure(figsize=(8,5))
sns.histplot(df['units_sold'], bins=30, kde=True, color="orange")
plt.title("Units Sold Distribution")
plt.show()

# --- Categorical features ---
print("\n--- Unique Value Counts for Categorical Columns ---")
for col in ['brand','category','channel','region','pack_type','promotion_flag']:
    print(f"\n{col}:\n", df[col].value_counts().head())

# Example: top 10 brands by frequency
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='brand', order=df['brand'].value_counts().index[:10])
plt.title("Top 10 Brands by Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
# 27. Bivariate Analysis

# Revenue by region
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='region', y='revenue', estimator=sum, ci=None)
plt.title("Total Revenue by Region")
plt.show()

# Revenue vs promotion
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='promotion_flag', y='revenue')
plt.title("Revenue Distribution: Promotion vs Non-Promotion")
plt.show()

# Correlation heatmap (numerical variables)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Example groupby
print("\n--- Average Revenue by Category ---")
print(df.groupby('category')['revenue'].mean().sort_values(ascending=False).head())
```

```python
# 28. Multivariate Analysis

# Pivot table: revenue by region & brand
pivot_table = df.pivot_table(values='revenue', index='region', columns='brand', aggfunc='sum')

plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap="YlGnBu")
plt.title("Revenue by Region & Brand")
plt.show()

# Pairplot for selected numeric features
sns.pairplot(df[['units_sold','price_unit','revenue']])
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()

# Multivariate groupby example
print("\n--- Revenue by Region, Category, and Promotion ---")
print(df.groupby(['region','category','promotion_flag'])['revenue'].sum().head(10))
```
